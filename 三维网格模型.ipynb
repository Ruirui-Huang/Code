```python
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator
from matplotlib.ticker import FuncFormatter
```

```python
path = r'D:\Matlab\Data\test.dat'
data = pd.read_table(path, sep = '\t', header=None, engine='python')
```

```python
model = np.zeros((32, 32, 16))
for i in range(len(data)):
    index = data.iloc[i]
    x = index[0] - 1
    y = index[1] - 1
    z = index[2] - 1
    model[x][y][z] = 1
model = model > 0

fig = plt.figure(figsize = (8, 4))
ax = fig.gca(projection='3d')
ax.voxels(model, facecolors='red', edgecolor='w', linewidth=0.5)


plt.xticks(np.arange(0, 33, 1), ('0', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '160', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '320', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '480', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '640'))
ax.set_xlabel('L(m)')
plt.yticks(np.arange(0, 33, 1), ('0', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '160', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '320', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '480', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '640'))
ax.set_ylabel('W(m)')
ax.set_zticks(np.arange(0, 17, 1))
ax.set_zticklabels(['-320', ' ', ' ', ' ', '-240 ', ' ', ' ', ' ', '-160', ' ', ' ', ' ', ' -80', ' ', ' ', ' ', '0'])
ax.set_zlabel('H(m)')

plt.show()
```
